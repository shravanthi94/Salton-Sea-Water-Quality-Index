```python
x = df['Year']
y1 = df['pH']
y2 = df['Temp']

# Plot Line1 (Left Y Axis)
fig, ax1 = plt.subplots(1,1,figsize=(16,9), dpi= 80)
ax1.plot(x, y1, color='tab:red')

# Plot Line2 (Right Y Axis)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(x, y2, color='tab:blue')

# Decorations
plt.xlim(2000,2010)
# ax1 (left Y axis)
ax1.set_xlabel('Year 2005', fontsize=20)
ax1.tick_params(axis='x', rotation=0, labelsize=12)
ax1.set_ylabel('pH', color='tab:red', fontsize=20)
ax1.tick_params(axis='y', rotation=0, labelcolor='tab:red' )
ax1.grid(alpha=.4)

# ax2 (right Y axis)
ax2.set_ylabel("Temperature", color='tab:blue', fontsize=20)
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.set_xticks(np.arange(0, len(x), 60))
ax2.set_xticklabels(x[::60], rotation=90, fontdict={'fontsize':10})
ax2.set_title("Data Distribution of Parameters: pH and Temperature", fontsize=22)
fig.tight_layout()
plt.show()
```

```python
# Prepare data
x_var = 'Temp'
groupby_var = 'Year'
df_agg = df.loc[:, [x_var, groupby_var]].groupby(groupby_var)
vals = [df[x_var].values.tolist() for i, df in df_agg]

# Draw
plt.figure(figsize=(12,8), dpi= 80)
colors = [plt.cm.Spectral(i/float(len(vals)-1)) for i in range(len(vals))]
n, bins, patches = plt.hist(vals, 30, stacked=True, density=False, color=colors[:len(vals)])

# Decoration
plt.legend({group:col for group, col in zip(np.unique(df[groupby_var]).tolist(), colors[:len(vals)])}, ncol = 2,loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Distribution of Temperature colored by Year", fontsize=22)
plt.xlabel(x_var, fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.ylim(0, 150)
plt.xticks(ticks=bins[::3], labels=[round(b,1) for b in bins[::3]])
plt.show()
```

```python
df['Year']= df['Date'].dt.year
df['Month']= df['Date'].dt.month

df_counts = df.groupby(['Year', 'Temp']).size().reset_index(name='counts')

# Stripplot plot
fig, ax = plt.subplots(figsize=(30,10), dpi= 80)    
sns.stripplot(df_counts.Temp, df_counts.Year, size=df_counts.counts*10, ax=ax)

plt.xlim(0, 40)
plt.title('Density Map: Distribution of Temperature in Raw Data (Size of the circle represents the density of data points)', fontsize=22)
plt.show()
```
