```python
import matplotlib.pyplot as plt 
TestDataAccuracy = ['98.89', '94.6', '98.89', '92.01','99.96']
TrainDataAccuracy = ['99.37','94.6', '99.00', '93.45', '98.68']
Loss = ['3.7215', '11.15', '20.6', '28.94', '17.7']
Models = ['RF', 'BNN', 'ANN', 'MLP', 'LSTM']
```

```python
x1 = Models 
y1 = Loss
# plotting the line 1 points  
plt.plot(x1, y1, label = "Loss") 

x2 = Models  
y2 = TrainDataAccuracy
# plotting the line 2 points  
plt.plot(x2, y2, label = "TrainDataAccuracy") 

x3 = Models  
y3 = TestDataAccuracy
# plotting the line 2 points  
plt.plot(x3, y3, label = "TestDataAccuracy") 

# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
# giving a title to my graph 
plt.title('Models Comparison Graph') 
  
# show a legend on the plot 
plt.legend() 
  
# function to show the plot 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = ['ANN','MLP', 'RF', 'BNN', 'LSTM']

plt.bar(X[0], 98.89, color = 'b', width = 0.25)
plt.bar(X[1], 92.01, color = 'g', width = 0.25)
plt.bar(X[2], 98.89, color = 'r', width = 0.25)
plt.bar(X[3], 94.6, color = 'b', width = 0.25)
plt.bar(X[4], 99.96, color = 'g', width = 0.25)
plt.title("Accuracy Comparison")
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.show()
```

```python
barWidth = 0.20
plt.figure(figsize=(10,7), dpi=80)

accuracy = [99.45,93.45,99.37,95.6,99.96]
test     = [98.89,92.01,98.89,94.6,99.96]
rmse     = [1.02,10.84,3.7,7,9]

r1 = np.arange(len(accuracy))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

plt.bar(r1, accuracy, color = 'b', width =barWidth, label='Accuracy of Train')
plt.bar(r2, test,  color = 'g', width = barWidth, label='Accuracy of Test')
plt.bar(r3, rmse, color = 'r', width = barWidth, label='RMSE')
plt.title("Model Comparison", fontweight='bold', fontsize=22)

# Add xticks on the middle of the group bars
plt.ylim(0,100)
plt.xticks([r + barWidth for r in range(len(accuracy))], ['Neural Networks','MLP', 'Random Forests', 'BNN', 'LSTM'])
plt.legend()
plt.show()
```
