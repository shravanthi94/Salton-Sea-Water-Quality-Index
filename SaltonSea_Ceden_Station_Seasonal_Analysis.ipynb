# Seasonal Analysis for complete data

```python
filename = 'Full_Data_Mean_Result.csv'
dfs = pd.read_csv(filename)
print(dfs.shape)
```

```python
dfs = dfs[['Date','pH', 'Salinity', 'Ammonia', 'Conductivity', 'DO', 'Temp']]
```

```python
dfs.head()
```

```python
dfs['Date'] = pd.to_datetime(dfs.Date)
```

```python
dfs['year'] = dfs['Date'].dt.year
dfs['month'] = dfs['Date'].dt.month
dfs['day'] = dfs['Date'].dt.day
```

```python
dfs.head()
```

```python
dfs = dfs.drop('Date', axis = 1)
```

```python
def getSeason(x):
    if x in range(1,4):
        return 0
    elif x in range(4,6):
        return 1
    elif x in range(6,9):
        return 2
    else:
        return 3
dfs['Season']=dfs['month'].apply(lambda x:getSeason(x))
dfs.head()
```

```python
x1 = dfs['Season'] 
y1 = dfs['pH']
# plotting the line 1 points  
plt.plot(x1, y1, label = "pH") 
  
# line 2 points 
x2 = dfs['Season']  
y2 = dfs['Salinity']
# plotting the line 2 points  
plt.plot(x2, y2, label = "Salinity") 

x3 = dfs['Season']  
y3 = dfs['Temp']
# plotting the line 2 points  
plt.plot(x3, y3, label = "Temp") 
  
x4 = dfs['Season']  
y4 = dfs['Conductivity']
# plotting the line 2 points  
plt.plot(x4, y4, label = "Conductivity") 

x5 = dfs['Season']  
y5 = dfs['Ammonia']
# plotting the line 2 points  
plt.plot(x5, y5, label = "Ammonia") 

x6 = dfs['Season']  
y6 = dfs['DO']
# plotting the line 2 points  
plt.plot(x6, y6, label = "DO") 

# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
# giving a title to my graph 
plt.title('Parameters distribution per Season') 
  
# show a legend on the plot 
plt.legend() 
  
# function to show the plot 
plt.show() 
```
